```python
#import Libraries

import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.preprocessing import LabelEncoder 
from sklearn.linear_model import LinearRegression 
from sklearn.model_selection import train_test_split

#Ignore all warnings 

import warnings
warnings.filterwarnings("ignore")
```

```python
# Load the Netflix Movies and TV Shows Data dataset

netflix_df = pd.read_csv("netflix_titles.csv")

# Display the first few rows of the dataset
netflix_df.head(5)
```

HANDLING MISSING VALUES 

```python
# Check for null values

null_values = netflix_df.isnull().sum()
null_values
```

Why Does It Matter?
When you have missing data, it can cause errors in analysis, visualization, or machine learning models. By filling in the missing data (called imputation), you make the dataset complete and ready for further analysis.

```python
# Fill null values in numerical columns with mean

netflix_df['director'].fillna(netflix_df['director'].mode()[0], inplace=True)
netflix_df['cast'].fillna(netflix_df['cast'].mode()[0], inplace=True)
netflix_df['country'].fillna(netflix_df['country'].mode()[0], inplace=True)
netflix_df['date_added'].fillna(netflix_df['date_added'].mode()[0], inplace=True)
netflix_df['duration'].fillna(netflix_df['duration'].mode()[0], inplace=True)

# Check if there are any remaining null values
null_values = netflix_df.isnull().sum()
null_values
```

DETECTING AND HANDLING OUTLIERS

```python
# Identify outliers using box plots

plt.figure(figsize=(10,6))
sns.boxplot(data=netflix_df, color='salmon')
plt.title('Identifying Outliers')
plt.show()
```

```python
# Handle outliers in release_year column

q1 = netflix_df['release_year'].quantile(0.5)
q3 = netflix_df['release_year'].quantile(0.75)

iqr = q3 - q1

lower_bound = q1 - 1.5 * iqr 
upper_bound = q3 + 1.5 * iqr 

netflix_df = netflix_df[(netflix_df['release_year'] >= lower_bound) & (netflix_df['release_year'] <= upper_bound)]
```

```python
# Convert release_year to numeric

netflix_df.dtypes
netflix_df['release_year'] = netflix_df['release_year'].astype(int)
```

EXPLORATORY DATA ANALYSIS 

```python
# Count the number of movies and TV shows
netflix_df.head()
```

```python
type_counts = netflix_df['type'].value_counts()

plt.figure(figsize=(8,6))
type_counts.plot(kind='bar', color='teal' )
plt.title("Distribution of Movie Types")
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
# Find the top directors

top_directors = netflix_df['director'].value_counts().head(10)
top_directors
```

```python
plt.figure(figsize=(10,8))
top_directors.plot(kind='bar', color='lightblue')
plt.title("Top 10 Directors")
plt.xlabel('Directors')
plt.ylabel('Count')
plt.show()
```

TOP CAST MEMBERS

```python
# Find the top cast members

top_cast = netflix_df['cast'].str.split(', ', expand=True).stack().value_counts().head(10)

plt.figure(figsize=(10,8))
top_cast.plot(kind='bar', color='purple')
plt.title("Top 10 Cast Members")
plt.xlabel('Cast Members')
plt.ylabel('Count')
plt.show()
```

MOVIE RELEASE YEARS 

```python
# Visualize the distribution of movie release years

plt.figure(figsize=(12,6))
sns.histplot(netflix_df['release_year'], bins=30, kde=True, color='teal')
plt.title('Distribution of Movie Release Years')
plt.xlabel('Release Year')
plt.ylabel('Count')
plt.show()
```

MOVIE RATINGS 

```python
# Visualize the distribution of movie ratings

plt.figure(figsize=(10,6))
netflix_df['rating'].value_counts().plot(kind='bar', color='lightgreen')
plt.title('Distribution of Movie Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```
