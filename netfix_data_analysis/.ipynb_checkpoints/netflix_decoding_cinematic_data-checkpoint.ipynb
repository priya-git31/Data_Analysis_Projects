```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
netflix_df = pd.read_csv('netflix.csv')
```

```python
netflix_df.head()
```

```python
netflix_df.describe()
```

```python
netflix_df.isnull().sum()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(netflix_df['imdb_score'], kde=True, color='teal')
plt.title("Distribution of IMDb scores")
plt.xlabel('Imdb Score')
plt.ylabel('Freuency')
plt.show()
```

```python
# Relationship between IMDb score and runtime
plt.figure(figsize=(10,6))
sns.scatterplot(data=netflix_df, x='runtime', y='imdb_score', color='red')
plt.title('ImDB Score VS Runtime')
plt.xlabel('Runtime(minutes)')
plt.ylabel('IMDb Score')
plt.show()
```

```python
# Relationship between IMDb score and runtime
plt.figure(figsize=(10,6))
sns.scatterplot(data=netflix_df, x='runtime', y='imdb_score', color='skyblue')
plt.title('Imdb Score VS Runtime')
plt.xlabel('Runtime(minutes)')
plt.ylabel('Imdb Score')
plt.show()
```

```python
length_of_data_set = len(netflix_df)
length_of_data_set
```

```python
# Top 10 languages by count

top_languages = netflix_df['language'].value_counts().head(10)
plt.figure(figsize=(10,6))
top_languages.plot(kind='bar', color='lightblue')
plt.title("Top 10 Languages")
plt.xlabel("Language")
plt.ylabel("Count")
plt.xticks(rotation=50)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.histplot(netflix_df['year'], bins=20, kde=True, color='Salmon')
plt.title('Release Year Distribution')
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.show()
```

```python
# Relationship between IMDb score and premiere year

plt.figure(figsize=(10,8))
sns.boxplot(data=netflix_df, x='year', y='imdb_score', palette='Set2')
plt.title("IMDb Score Distribution by Premiere Year")
plt.xlabel("Year")
plt.ylabel("IMDb Score")
plt.xticks(rotation=90)
plt.show()
```

```python
# Runtime distribution

plt.figure(figsize=(10,6))
sns.histplot(netflix_df['runtime'], bins=20, kde=True, color='teal')
plt.title("Runtime Distribution")
plt.xlabel("Runtime(minutes)")
plt.ylabel("Frequency")
plt.show()
```

```python
# IMDb score distribution over time
plt.figure(figsize=(12, 8))
sns.lineplot(data=netflix_df, x='year', y='imdb_score', color='orange')
plt.title('IMDb Score Trend Over Time')
plt.xlabel('Year')
plt.ylabel('IMDb Score')
plt.xticks(rotation=90)
plt.show()
```

```python
# Count of titles by premiere year

plt.figure(figsize=(10,8))
sns.countplot(data=netflix_df, x='year', palette='rocket')
plt.title("Count of Titles by Premiere Year")
plt.xlabel("Year")
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
# Relationship between IMDb score and language for top 5 languages
top_5_languages = netflix_df['language'].value_counts().head(5).index
top_5_data = netflix_df[netflix_df['language'].isin(top_5_languages)]
plt.figure(figsize=(12, 8))
sns.boxplot(data=top_5_data, x='language', y='imdb_score', palette='muted')
plt.title('IMDb Score Distribution by Top 5 Languages')
plt.xlabel('Language')
plt.ylabel('IMDb Score')
plt.xticks(rotation=45)
plt.show()
```
