# Predicting Whether a Customer Will Purchase (Yes / No)

### Business Context 

An e-commerce company wants to predict whether a customer will make a purchase based on:
1. How many minutes they spent on the website
2. How many pages they viewed
3. Whether they came via an ad campaign
4. Their age

This is a binary classification problem:

1 = Customer purchased

0 =  Customer did NOT purchase

This makes it a  Logistic Regression Problem

### Objective

Build a Logistic Regression model that predicts:
Will the customer purchase? (Yes = 1 / No = 0)

### Step 1: Create the Dataset

```python
import pandas as pd

# Create dataset
data = {
    'time_on_site': [5, 10, 3, 20, 15, 2, 25, 18, 7, 30],
    'pages_viewed': [2, 5, 1, 10, 8, 1, 12, 9, 3, 15],
    'ad_click':     [0, 1, 0, 1, 1, 0, 1, 1, 0, 1],
    'age':          [22, 35, 19, 45, 38, 21, 50, 42, 25, 55],
    'purchased':    [0, 1, 0, 1, 1, 0, 1, 1, 0, 1]
}

df = pd.DataFrame(data)
df
```

### Step 2: Understand the Data

```python
df.info()
```

```python
df.describe()
```

### Step 3: Define X (features) and y (target)
### Step 4: Train–Test Split

```python
X = df[['time_on_site', 'pages_viewed', 'ad_click', 'age']]
y = df['purchased']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

### Step 5: Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

### Step 6: Train Logistic Regression Model

```python

from sklearn.linear_model import LogisticRegression 

model = LogisticRegression()
model.fit(X_train_scaled, y_train)
```

### Step 7: Make Predictions

```python
# 1️⃣ Class predictions (0 or 1)

y_pred = model.predict(X_test_scaled)
y_pred
```

```python
# 2️⃣ Probability predictions (THIS IS LOGISTIC MAGIC)

y_prob = model.predict_proba(X_test_scaled)
y_prob
```

### Step 8: Model Evaluation

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, y_pred)
```

### Accuracy

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```

```python
# Classification Report

from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

### Step 9: Understand the Model Coefficients

```python
coefficients = pd.DataFrame({
    'Feature': X.columns,
    'Coefficient': model.coef_[0]
})

coefficients
```

How to interpret:

Positive coefficient → increases probability of purchase

Negative coefficient → decreases probability

Example:

If time_on_site coefficient = +1.2
→ More time on site → higher chance of purchase

### Step 10: Interpret Using Odds

```python
import numpy as np

coefficients['Odds_Ratio'] = np.exp(coefficients['Coefficient'])
coefficients
```

Interpretation:

Odds Ratio > 1 → increases odds

Odds Ratio < 1 → decreases odds

### Step 11: Predict for a New Customer

```python
new_customer = pd.DataFrame({
    'time_on_site': [12],
    'pages_viewed': [6],
    'ad_click': [1],
    'age': [32]
})

new_customer_scaled = scaler.transform(new_customer)

model.predict(new_customer_scaled)
model.predict_proba(new_customer_scaled)
```
